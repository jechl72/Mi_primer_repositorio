```python
# Parte 2 - Punto 2: Calculadora de tiempo

# 1) Asignar a una variable llamada segundos_por_hora las operaciones correspondientes para calcular dicho valor.
segundos_por_hora = 60 * 60  # 60 segundos por minuto * 60 minutos por hora

# 2) Calcular cuántos segundos hay en un día y asignarlo a la variable segundos_por_dia.
segundos_por_dia = segundos_por_hora * 24  # 24 horas por día

print("\n=== Calculadora de tiempo ===")
print(f"Segundos por hora: {segundos_por_hora}")
print(f"Segundos por día: {segundos_por_dia}")

# 3) Divide segundos_por_dia entre segundos_por_hora usando división con punto flotante (/).
print("\nDivisión con punto flotante (/):")
print(segundos_por_dia / segundos_por_hora)

# 4) Divide segundos_por_dia entre segundos_por_hora usando división entera (//).
print("\nDivisión entera (//):")
print(segundos_por_dia // segundos_por_hora)

# 5) Análisis breve de resultados
print("\n=== Análisis ===")
print("Ambas divisiones deben dar 24, porque un día tiene 24 horas.")
print("Con / el resultado es tipo float (puede mostrarse como 24.0).")
print("Con // el resultado es tipo int (se muestra como 24).")
```
